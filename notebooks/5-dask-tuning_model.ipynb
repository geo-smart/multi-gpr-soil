# Pytorch Multi-output GPR

The purpose of this notebook is to improve the predictions of Layer 1 soil type and soil thickness of Seattle from the preliminary work demostrated in notebook `4-pytorch_gpr.ipynb`.

The tactics used to improve the model include: 
1. Performing a log scale transformation of the layer thickness target (instead of a standard scaler).
2. Evaluating class imbalance of the classifier. 
3. Performing hyperparameter tuning, looking at different kernels and learning rates.
4. Performing cross-validation.

```python
class GPClassificationModel(ApproximateGP):
    def __init__(self, train_x, num_classes):
        variational_distribution = CholeskyVariationalDistribution(train_x.size(0))
        variational_strategy = VariationalStrategy(self, train_x, variational_distribution, learn_inducing_locations=True)
        super(GPClassificationModel, self).__init__(variational_strategy)
        
        self.mean_module = gpytorch.means.ConstantMean()
        self.covar_module = gpytorch.kernels.ScaleKernel(gpytorch.kernels.RBFKernel())
        self.num_classes = num_classes

    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x).expand([self.num_classes])
```

```python
class GPRegressionModel(ExactGP):
    def __init__(self, train_x, train_y, likelihood):
        super(GPRegressionModel, self).__init__(train_x, train_y, likelihood)
        self.mean_module = gpytorch.means.ConstantMean()
        self.covar_module = ScaleKernel(RBFKernel())

    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return MultivariateNormal(mean_x, covar_x)
```

```python
@dask.delayed
def dask_hyperparam_tune(model_class, X, y, lr, lengthscale, outputscale, kf):
    fold_scores = []
        
    for train_index, val_index in kf.split(X):
        # print('running:',train_index, val_index)
        X_train_fold, X_val_fold = X[train_index], X[val_index]
        y_train_fold, y_val_fold = y[train_index], y[val_index]

        # Initialize model with specific hyperparameters
        likelihood = gpytorch.likelihoods.GaussianLikelihood()
        model = model_class(X_train_fold, y_train_fold, likelihood)
        
        # Set hyperparameters
        model.covar_module.base_kernel.lengthscale = lengthscale
        model.covar_module.outputscale = outputscale
        
        # Training
        model.train()
        likelihood.train()

        optimizer = torch.optim.Adam(model.parameters(), lr=lr)
        mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)
        
        training_iter = 50
        for i in range(training_iter):
            optimizer.zero_grad()
            output = model(X_train_fold)
            loss = -mll(output, y_train_fold)
            loss.backward()
            optimizer.step()
        
        # Validation
        model.eval()
        likelihood.eval()
        with torch.no_grad(), gpytorch.settings.fast_pred_var():
            val_output = model(X_val_fold)
            val_loss = -mll(val_output, y_val_fold)
            fold_scores.append(val_loss.item())
    
        avg_score = np.mean(fold_scores)

    out = xr.DataArray(
                        coords={'avg_score': avg_score},
                        vars={'lr': lr, 
                              'lengthscale': lengthscale, 
                              'outputscale': outputscale, 
                              'model': model, 
                              'likelihood': likelihood},
                        )
    
    print(out)
    return(out)
    # return(avg_score, lr, lengthscale, outputscale, model, likelihood)#{'avg_score': avg_score, 'lr': lr, 'lengthscale': lengthscale, 'outputscale': outputscale, 'model': model, 'likelihood': likelihood})

def cross_val_score_gp(model_class, X, y, param_grid, cv=5):
    kf = KFold(n_splits=cv)
    # best_score = float('inf')
    # best_params = None
    score_params = []

    for lr in param_grid['lr']:
        for lengthscale in param_grid['lengthscale']:
            for outputscale in param_grid['outputscale']:
                # avg_score, lr, lengthscale, outputscale, model, likelihood = dask_hyperparam_tune(X, y, lr, lengthscale, outputscale)
                # print(avg_score, lr, lengthscale, outputscale, model, likelihood)
                hyper_scores = dask_hyperparam_tune(model_class, X, y, lr, lengthscale, outputscale, kf)
                # hyper_scores = xr.DataArray(
                #                             coords={'avg_score': avg_score},
                #                             vars={'lr': lr, 
                #                                   'lengthscale': lengthscale, 
                #                                   'outputscale': outputscale, 
                #                                   'model': model, 
                #                                   'likelihood': likelihood},
                                            # )
                # print(hyper_scores)
                score_params.append(hyper_scores)
                # fold_scores = []
                
                # for train_index, val_index in kf.split(X):
                    # # print('running:',train_index, val_index)
                    # X_train_fold, X_val_fold = X[train_index], X[val_index]
                    # y_train_fold, y_val_fold = y[train_index], y[val_index]

                    # # Initialize model with specific hyperparameters
                    # likelihood = gpytorch.likelihoods.GaussianLikelihood()
                    # model = model_class(X_train_fold, y_train_fold, likelihood)
                    
                    # # Set hyperparameters
                    # model.covar_module.base_kernel.lengthscale = lengthscale
                    # model.covar_module.outputscale = outputscale
                    
                    # # Training
                    # model.train()
                    # likelihood.train()

                    # optimizer = torch.optim.Adam(model.parameters(), lr=lr)
                    # mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)
                    
                    # training_iter = 50
                    # for i in range(training_iter):
                    #     optimizer.zero_grad()
                    #     output = model(X_train_fold)
                    #     loss = -mll(output, y_train_fold)
                    #     loss.backward()
                    #     optimizer.step()
                    
                    # # Validation
                    # model.eval()
                    # likelihood.eval()
                    # with torch.no_grad(), gpytorch.settings.fast_pred_var():
                    #     val_output = model(X_val_fold)
                    #     val_loss = -mll(val_output, y_val_fold)
                    #     fold_scores.append(val_loss.item())

                # avg_score = np.mean(fold_scores)
                
                # if avg_score < best_score:
                #     best_score = avg_score
                #     best_params = params
    
    #Let's find the best of the best
    with dask.diagnostics.ProgressBar():
        print('Dask scoring')
        score_params = dask.compute(score_params)
    # score_params_comp = []
    # for i in score_params:
    #     i.visualize
    #     score_params_comp.append(i.compute())
    score_params = xr.concat(score_params, dim='avg_score')
    
    best_score = score_params.max(dim='avg_score').values
    best_params = {'lengthscale': score_params.lengthscale.sel(avg_score=best_score).values, 
                   'outputscale': score_params.outputscale.sel(avg_score=best_score).values}
    model = score_params.model.sel(avg_score=best_score).values
    likelihood = score_params.likelihood.sel(avg_score=best_score).values
    print(f'The best score found was {best_score}.')

    return best_params, best_score, model, likelihood
```
